```python
df1=pd.read_excel("Country-Code.xlsx")
```

```python
df1
```

```python
dfn=pd.merge(df,df1,on='Country Code',how='left')
```

```python
dfn
```

```python
dfn.dtypes
```

```python
country_name=dfn.Country.value_counts().index
```

```python
country_values=dfn.Country.value_counts().values
```

```python
plt.pie(country_values[:3],labels=country_name[:3],autopct="%1.2f%%")
plt.show()
```

```python
#observation:zomato maximum records and transaction from india
```

```python
dfn.columns
```

```python
rating=dfn.groupby([ 'Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
rating
```

# observation:
4.5  to 4.9=Excellent,
4.0  to 4.4=v good
3.5  to 4.9=good
2.5  to 3.4=avg
1.8  to 2.4=poor
0.0= not rated

```python
rating.head()
```

```python
plt.rcParams['figure.figsize']=(12,6)
sns.barplot(x="Aggregate rating",y="Rating Count",hue='Rating color',data=rating,palette=['pink','red','orange','yellow',"green",'black'])
```

```python
#observation not rated count is very high
#max no of rating are between 2.5,3.4
```
